# 3.Calificación promedio de la tienda

En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es **conocer la satisfacción del cliente con los productos vendidos**.

```python
#   Promedio de satisfacción del cliente con los productos vendidos de Tienda 1
print('Tienda 1')
cant_calific_1 = tienda1['Calificación'].count()
print('Cantidad de calificaciones:', cant_calific_1)
suma_calific_1 = tienda1['Calificación'].sum()
print('Suma total de calificaciones:',suma_calific_1)
promedio_calific_1 = suma_calific_1 / cant_calific_1
print(f'Calificación promedio de Tienda 1 es de {promedio_calific_1:,.2f}')
```

```python
#   Promedio de satisfacción del cliente con los productos vendidos de Tienda 2
print('Tienda 2')
cant_calific_2 = tienda2['Calificación'].count()
print('Cantidad de calificaciones:', cant_calific_2)
suma_calific_2 = tienda2['Calificación'].sum()
print('Suma total de calificaciones:',suma_calific_2)
promedio_calific_2 = suma_calific_2 / cant_calific_2
print(f'Calificación promedio de Tienda 2 es de {promedio_calific_2:,.2f}')
```

```python
#   Promedio de satisfacción del cliente con los productos vendidos de Tienda 3
print('Tienda 3')
cant_calific_3 = tienda3['Calificación'].count()
print('Cantidad de calificaciones:', cant_calific_3)
suma_calific_3 = tienda3['Calificación'].sum()
print('Suma total de calificaciones:',suma_calific_3)
promedio_calific_3 = suma_calific_3 / cant_calific_3
print(f'Calificación promedio de Tienda 3 es de {promedio_calific_3:,.2f}')
```

```python
#   Promedio de satisfacción del cliente con los productos vendidos de Tienda 4
print('Tienda 4')
cant_calific_4 = tienda4['Calificación'].count()
print('Cantidad de calificaciones:', cant_calific_4)
suma_calific_4 = tienda4['Calificación'].sum()
print('Suma total de calificaciones:',suma_calific_4)
promedio_calific_4 = suma_calific_4 / cant_calific_4
print(f'Calificación promedio de Tienda 4 es de {promedio_calific_4:,.2f}')
```

```python
#   Promedio de satisfacción del cliente con los productos vendidos en las 4 tiendas
#   Usando .mean  para  corroborrar

def calificacion_tienda(tienda, numero):
    promedio = tienda['Calificación'].mean()
    print(f"Calificación promedio en Tienda {numero}: {promedio:.2f}")

# Generando resultados de cada tienda:
calificacion_tienda(tienda1, 1)
calificacion_tienda(tienda2, 2)
calificacion_tienda(tienda3, 3)
calificacion_tienda(tienda4, 4)
```

Graficar tendencia de ventas mensuales:

```python
# convertir fechas a  datetime
tienda1['Fecha de Compra'] = pd.to_datetime(tienda1['Fecha de Compra'], dayfirst=True)
tienda2['Fecha de Compra'] = pd.to_datetime(tienda2['Fecha de Compra'], dayfirst=True)
tienda3['Fecha de Compra'] = pd.to_datetime(tienda3['Fecha de Compra'], dayfirst=True)
tienda4['Fecha de Compra'] = pd.to_datetime(tienda4['Fecha de Compra'], dayfirst=True)
```

```python
# Calcular ventas mensuales en cada tienda
# Agrupar por fecha y sumar precio
venta_mensual_1 = tienda1.groupby([tienda1['Fecha de Compra'].dt.to_period('M')])['Precio'].sum().reset_index()
venta_mensual_2 = tienda1.groupby([tienda2['Fecha de Compra'].dt.to_period('M')])['Precio'].sum().reset_index()
venta_mensual_3 = tienda1.groupby([tienda3['Fecha de Compra'].dt.to_period('M')])['Precio'].sum().reset_index()
venta_mensual_4 = tienda1.groupby([tienda4['Fecha de Compra'].dt.to_period('M')])['Precio'].sum().reset_index()
venta_mensual_3.head(1)
```

```python
#  Crear columna Tienda ID
venta_mensual_1['Tienda_ID'] = 'Tienda 1'
venta_mensual_2['Tienda_ID'] = 'Tienda 2'
venta_mensual_3['Tienda_ID'] = 'Tienda 3'
venta_mensual_4['Tienda_ID'] = 'Tienda 4'
#
hd_todas = pd.concat([venta_mensual_1,venta_mensual_2,venta_mensual_3,venta_mensual_4], ignore_index=True)
hd_todas.head(2)
```

```python
#  Vamos a crea tabla  ventas_mensuales  de nuestro DataFrame  meses
#  Que tenga los meses x las tiendas.  Y en datos, la cantidad vendida de productos
#  Usamos el  picot_table  para crear una tabla dinámica
#  El índice son los  meses
#  Las cols  son las tiendas
#  Los valores son la cantidad vendida de productos
#  sum() es la función de agregación
#  sort = False     para que NO  se reordene al crear la tabla.
ventas_mensuales = hd_todas.pivot_table(index=hd_todas.Tienda_ID, columns='Fecha de Compra', values='Precio', aggfunc='sum', sort=False)
ventas_mensuales = ventas_mensuales.T
ventas_mensuales
```

```python
#  tendencia de ventas a lo largo de los meses de cada tienda.
ventas_mensuales.plot(ylabel='Ventas',title = 'Ventas mensuales',figsize=(12, 6));

plt.title('Tendencia de ventas por tienda', fontsize=15)
plt.ylabel('Monto de ventas Mensuales (USD)')
#plt.xlabel('Mes')

plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Ventas menuales de cada tienda:
venta_mensual_1.plot(xlabel='Tiempo',ylabel='Ventas',title = 'Ventas mensuales Tienda 1');
venta_mensual_2.plot(xlabel='Tiempo',ylabel='Ventas',title = 'Ventas mensuales Tienda 2');
venta_mensual_3.plot(xlabel='Tiempo',ylabel='Ventas',title = 'Ventas mensuales Tienda 3');
venta_mensual_4.plot(xlabel='Tiempo',ylabel='Ventas',title = 'Ventas mensuales Tienda 4');
```
